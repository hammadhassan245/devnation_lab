```python
import pandas as pd
```

```python
df =pd.read_csv('olympics.csv', index_col=0,skiprows=1)
for col in df.columns:
    if col[:2] =='01':
        df.rename(columns={col:'Gold'+col[4:]}, inplace=True)
    if col[:2] =='02':
        df.rename(columns={col:'Silver'+col[4:]}, inplace=True)
    if col[:2] =='03':
        df.rename(columns={col:'Brozne'+col[4:]}, inplace=True)
names_ids=df.index.str.split('\s\(')
df.index=names_ids.str[0]
df['ID'] =names_ids.str[1].str[:3]
df=df.drop('Totals')
df.head()
```

```python
#question no 01

country = df.iloc[0]
country
```

```python
#Question no 02

find_max = df["Gold"].max()
summer = df.loc[df["Gold"] == find_max]
summer

```

```python
#Question no 03
raw = df.loc[(df["Gold"] != 0)&(df["Gold.1"] != 0)]
diff = ((raw["Gold"]-raw['Gold.1'])/raw['Gold.2']).max()
diff
```

```python
#Question no 04

df['point'] = (df["Gold.2"]*3)+(df["Silver.2"]*2)+(df["Brozne.2"]*1)
df.head()
```

```python
#Question no 05
df =pd.read_csv('census.csv', index_col=0)
raw = df.groupby('STNAME')['COUNTY'].agg(['sum'])
# raw = df.groupby('STNAME')['COUNTY'].agg(['sum']).reset_index()
find_max = raw["sum"].max()
mostcounties = raw.loc[raw["sum"] == find_max]
mostcounties
```

```python
#Question no 06
mostpop = df.groupby('STNAME')['CENSUS2010POP'].nlargest(3).reset_index()
mostpop=mostpop.drop('SUMLEV',axis=1)
mostpop
```

```python
#Question no 07

newdataset = df[['POPESTIMATE2010', 'POPESTIMATE2011', 'POPESTIMATE2012','POPESTIMATE2013','POPESTIMATE2014','POPESTIMATE2015']]
largest = newdataset.max(axis=1)
smallest=newdataset.min(axis=1)

df['diff']=(largest - smallest)

find_max = df["diff"].max()
fetch = df.loc[df["diff"] == find_max]
fetch
```
